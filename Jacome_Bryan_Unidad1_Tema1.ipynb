# **1. Diseño de la clase**

Cree una clase llamada AnalizadorDatos con los siguientes elementos:

• Atributo: datos (una lista de números).

• Métodos:

* obtener_maximo() → devuelve el valor máximo.

* obtener_minimo() → devuelve el valor mínimo.

* calcular_media() → devuelve el promedio de los datos.

* mostrar_resumen() → imprime todos los valores anteriores.



```python
# Definición de la clase AnalizadorDatos

class AnalizadorDatos:

    # Constructor: recibe una lista de números
    def __init__(self, datos):
        self.datos = datos

    # Método para obtener el valor máximo
    def obtener_maximo(self):
        return max(self.datos)

    # Método para obtener el valor mínimo
    def obtener_minimo(self):
        return min(self.datos)

    # Método para calcular el promedio
    def calcular_media(self):
        return sum(self.datos) / len(self.datos)

    # Método para mostrar el resumen
    def mostrar_resumen(self):
        print("Máximo:", self.obtener_maximo())
        print("Mínimo:", self.obtener_minimo())
        print("Media:", self.calcular_media())
```

#**3. Extensión: lectura dinámica**

a) Añada a la clase un método agregar_dato(nuevo_valor) que
permita añadir un nuevo número a la lista.

b) Permita que el usuario ingrese varios valores por teclado hasta
escribir “fin”.

c) Luego, muestre nuevamente el resumen con los nuevos datos
agregados

```python

# a) Añada a la clase un metodo agregar_dato(nuevo_valor) que permita añadir un
# nuevo numero a la lista.
class AnalizadorDatos:

    def __init__(self, datos):
        self.datos = datos

    def obtener_maximo(self):
        return max(self.datos)

    def obtener_minimo(self):
        return min(self.datos)

    def calcular_media(self):
        return sum(self.datos) / len(self.datos)

   # para añadir un nuevo dato
    def agregar_dato(self, nuevo_valor):
        self.datos.append(nuevo_valor)

    def mostrar_resumen(self):
        print("Máximo:", self.obtener_maximo())
        print("Mínimo:", self.obtener_minimo())
        print("Media:", self.calcular_media())
        print("Rango:", self.obtener_maximo() - self.obtener_minimo())

```

```python
#b) Permita que el usuario ingrese varios valores por teclado hasta escribir
# fin.

while True:
    entrada = input("Ingrese un número (o 'fin' para terminar): ")

    if entrada.lower() == "fin":
        break

    try:
        numero = float(entrada)
        analizador.agregar_dato(numero)
    except:
        print("Entrada inválida")
```

```python
# c)Luego, muestre nuevamente el resumen con los nuevos datos agregados
print("Cantidad de datos:", len(analizador.datos))
print("Datos actualizados:", analizador.datos)
print("Resumen actualizado:")
analizador.mostrar_resumen()

```

* *¿Cómo ayuda la orientación a objetos a organizar este tipo de
programas?*

La orientación a objetos nos ayuda a organizar  mejor el programa porque permite agrupar los datos y las funciones en una sola clase.

Esto nos permite que el codigo  sea más ordenado, fácil de entender y reutilizable.


* *¿Qué ventajas tiene encapsular las operaciones estadísticas dentro
de una clase?*


Encapsular las operaciones estadísticas dentro de una clase nos permite organizar mejor el código y evitar repeticiones inecesarias.

Además, nos facilita reutilizar las funciones y trabajar con los datos de manera más ordenada y tener una mejor eficiencia.
